## Hypotheses

- Science $H_0$: Aggressiveness of $Verticillium\;dahliae$ towards potato is not attenuated after serial passage through brown mustard and barley.
              
- Science $H_0$: Asymptomatic infections are stable after serial passage though mustard and barley.            

- Statistical $H_0$: $Y_{ijk} = u + E_{ijk}$ vs. $H_A$: $Y_{ijk} = u + A_i + B_j + C_k + E_{ijk}$

- Observed variation in response variables represent random variation, not systematic effects of hosts and isolates. Variation in the response variable is unrelated to variation in the predictor variable and is no greater than expected by chance or sampling error.

## Materials and methods: serial inoculations

$G_1$

- Inoculum preparation for soil drench in 3.5" pots:
    - $10^5 = \frac{100,000 CFU}{cm^3} \cdot \frac{500 cm^3}{pot} = \frac{50,000,000 CFU}{pot}$
    - $\frac{50,000,000 CFU}{pot} \cdot \frac{1 mL}{10^6 CFU} = \frac{50mL}{pot}$
- All isolates
    - inoculate with $\frac{50mL}{pot}$ of $\frac{10^6 CFU}{mL}$
- Total inoculum for all isolates
    - $\frac{50mL}{pot} \cdot 3$ hosts $\cdot 10 $ replicates $\cdot 2 $ trials $= \frac{3000mL}{isolate}$

$G_2, G_3, and\;G_4$

- Inoculum preparation for soil drench in 3.5" pots:
    - $10^5 = \frac{100,000 CFU}{cm^3} \cdot \frac{500 cm^3}{pot} = \frac{50,000,000 CFU}{pot}$
    - $\frac{50,000,000 CFU}{pot} \cdot \frac{1 mL}{10^6 CFU} = \frac{50mL}{pot}$
- All isolates
    - inoculate with $\frac{50mL}{pot}$ of $\frac{10^6 CFU}{mL}$
- Total inoculum for all isolates
    - $\frac{50mL}{pot} \cdot 1$ hosts $\cdot 10 $ replicates $\cdot 2 $ trials $= \frac{500mL}{isolate}$
- Inoculum for each isolate
    - $\frac{\frac{500mL}{isolate}}{\frac{200mL}{culture}} = 2.5 = 3$ cultures/isolate/trial or 5 cultures/isolate

## Materials and methods: bioassay

- **Inoculum preparation for soil drench in 3.5" pots with 10 replicates:**
    - $10^5 = \frac{100,000 CFU}{cm^3} \cdot \frac{500 cm^3}{pot} = \frac{50,000,000 CFU}{pot}$
    - $\frac{50,000,000 CFU}{pot} \cdot \frac{1 mL}{10^6 CFU} = \frac{50mL}{pot}$
- All isolates
    - inoculate with $\frac{50mL}{pot}$ of $\frac{10^6 CFU}{mL}$
- Total inoculum for all isolates
    - $\frac{50mL}{pot} \cdot 1$ host $\cdot 10 $ replicates $\cdot 2 $ trials $= \frac{1000mL}{isolate}$

- Inoculum for each isolate
    - $\frac{\frac{500mL}{isolate}}{\frac{200mL}{culture}} = 2.5 = 3$ cultures/isolate/trial or 5 cultures/isolate
    - $\frac{5 cultures}{isolate} \cdot 36\; isolates = 180\;cultures$
    - $\frac{3 cultures}{isolate} \cdot 36\; isolates = 108\;cultures$

- **Inoculum preparation for soil drench in 3.5" pots with 5 replicates:**
    - $10^5 = \frac{100,000 CFU}{cm^3} \cdot \frac{500 cm^3}{pot} = \frac{50,000,000 CFU}{pot}$
    - $\frac{50,000,000 CFU}{pot} \cdot \frac{1 mL}{10^6 CFU} = \frac{50mL}{pot}$
- All isolates
    - inoculate with $\frac{50mL}{pot}$ of $\frac{10^6 CFU}{mL}$
- Total inoculum for all isolates
    - $\frac{50mL}{pot} \cdot 1$ host $\cdot 5 $ replicates $\cdot 2 $ trials $= \frac{500mL}{isolate}$

- Inoculum for each isolate
    - $\frac{\frac{250mL}{isolate}}{\frac{200mL}{culture}} = 1.5 = 2$ cultures/isolate/trial or 4 cultures/isolate
    - $\frac{4 cultures}{isolate} \cdot 36\; isolates = 144\;cultures$
    - $\frac{2 cultures}{isolate} \cdot 36\; isolates = 72\;cultures$
